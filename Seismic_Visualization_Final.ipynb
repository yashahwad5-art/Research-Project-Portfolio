```python
plt.imshow(cube[500,:,:], cmap="seismic", aspect="auto")
plt.title("Inline 500")
plt.colorbar()
plt.show()

plt.imshow(cube[:,250,:], cmap="seismic", aspect="auto")
plt.title("Crossline 250")
plt.colorbar()
plt.show()

plt.imshow(cube[:,:,200], cmap="seismic")
plt.title("Time slice 200")
plt.colorbar()
plt.show()
```

```python
vol = cube[::3, ::3, ::2]
print(vol.shape)
```

```python
grid = pv.ImageData()
grid.dimensions = vol.shape
grid.spacing = (1, 1, 1)   # relative spacing
grid.point_data["Amplitude"] = vol.flatten(order="F")
```

```python
p = pv.Plotter()
p.add_volume(grid, cmap="seismic")
p.show()
```
